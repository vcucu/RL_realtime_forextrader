```python
import numpy as np
import pandas as pd
```

```python
filename = "testing_resultsRRL3.csv"
```

```python
df = pd.read_csv(filename)
```

```python
df['Current_val'] = df['Current_val'] - 98000
print(df['Current_val'].max())
print(df['Current_val'].min())
```

```python
df[['Time','Current_val']].plot(kind="line", x = 'Time', y = 'Current_val');
```

```python
#df["returns"] = df['Current_val'].apply(lambda x)
df['diff'] = df.groupby('Time')['Current_val'].diff().fillna(0)
df['diff_mean'] = df['diff'].rolling(window=20,axis=0).mean().fillna(0)
df['diff_std'] = df['diff'].rolling(window=20,axis=0).std().fillna(0)
df['diff_relative'] = df['diff']/df['Current_val']
df[:30]
```

```python
mean = df['diff_relative'].mean()
#-0.015926309948555987
std = df['diff_relative'].std()
#0.17984899747846084
```
